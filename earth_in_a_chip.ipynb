## *What "DNA" and "Proteins" Symbolise*


---



The way I'm about to explain this, it's mostly my thought train going along, so forgive me for any clarity issues:

So, what I am calling "DNA" and "Protein" are not full one-to-one simulation of how DNA and proteins are systhesized. The aim of Earth-in-a-chip is to simulate ecosystems, mutations, and evolution with as little hard-coding as possible.

So, what exactly do "DNA" and "Protein" mean in this project's context. Well, the simplest definition of (actual) DNA that I can think of is a recipe for how to contruct proteins, and that's oversimplifying it, but it will work for my explanation.

The dna object, in the code, is a list. It contains (direct) information on the traits of the organism (which is an Organism() Python class) unlike actual DNA, which create protein, which in turn create enzymes, which create other biological stuff, which let's the organism do stuff (functions) and get a structure (I'll come to structure in a second).

"Proteins", in this context, are plain Python functions. They take stuff in, do stuff, and produce in simulation side effects. The weights/values of these so-called "proteins" change by factors such as exposure to radiation, and a tiny bit of randomness, copying errors in DNA copying.






```python
def main():

  pygame.init()

  clock = pygame.time.Clock()
  FPS = 60

  running = True

  ###--The Infinite Loop--###

  while running:
    for event in pygame.event.get():
        if event.type == pygame.QUIT:
            running = False

    clock.tick(FPS)

  ###----###


  pygame.quit()
  sys.exit()
```
