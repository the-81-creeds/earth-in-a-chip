```python
class World():

  def __init__(self, xSize, ySize) -> None:
    self.worldBlocks = []
    self.worldArea = xSize * ySize

    for coordinate in range(self.worldArea):
      self.worldBlocks.append({'elevation': 0.0,

                               'light': 0.0,

                               'o2': 0.0,
                               'co2': 0.0,
                               'no2': 0.0,
                               'so2': 0.0,
                               'water': 0.0,

                               'permeability': 0.0})
    self.organisms = []

  def spawn(self, organism):
    self.organisms.append(organism)

class Organism():

  def __init__(self, id, world: World, x, y, dna: dict) -> None:
    self.id = id
    self.world = world

```
