```python
import pygame
import sys
```

Pygame is needed for the infinite loop and also for the future graphical simulation of virtual environment simulations

```python


class World():

  def __init__(self, xSize: int, ySize: int) -> None:
    self.worldBlocks = []
    self.worldArea = xSize * ySize

    for coordinate in range(self.worldArea):
      self.worldBlocks.append({'elevation': 0.0,

                               'light': 0.0,

                               'o2': 0.0,
                               'co2': 0.0,
                               'no2': 0.0,
                               'so2': 0.0,
                               'water': 0.0,

                               'radioactivity': 0.0,

                               'permeability': 0.0})

  ###--The Index Coordinate Functions--###

  def getIndex(self, x: int, y: int):
    return (y * self.xSize) + x

  def getX(self, index: int):
    return index % self.xSize

  def getY(self, index: int):
    return index // self.xSize

  ###----####

  def spawn(self, organism, x: int, y: int):
    index = self.getIndex(x, y)
    self.worldBlocks[index]['organism'] = organism

```

```python
class Organism():

  def __init__(self, id: int, world: World, x: int, y: int, dna: dict, nutrients: dict) -> None:

    self.id = id

    self.ammonia = nutrients['ammonia']
    self.phosphate = nutrients['phosphate']
    self.glucose = nutrients['glucose']




    world.spawn(self, x, y)
```

```python
def main():

  pygame.init()

  clock = pygame.time.Clock()
  FPS = 60

  running = True

  ###--The Infinite Loop--###

  while running:
    for event in pygame.event.get():
        if event.type == pygame.QUIT:
            running = False

    clock.tick(FPS)



  ###----###


  pygame.quit()
  sys.exit()
```
